```python
import numpy as np #mengelola data numerik
import matplotlib.pyplot as plt #untuk membuat grafik
import pandas as pd #mengelola data excel
```

```python
dataset = pd.read_csv('Used_Car_Purchase_Decision.csv')
x = dataset.iloc[:, [2,3]].values
y = dataset.iloc[:, -1].values
```

```python
print (x)
```

```python
print (y)
```

```python
from sklearn.model_selection import train_test_split #bagi datanya menggunakan library sklearn
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0)
```

```python
print (x_train)
```

```python
print (y_train)
```

```python
print (x_test)
```

```python
print (y_test)
```

```python
from sklearn.preprocessing import StandardScaler #mengubah skala data
sc_x = StandardScaler()
x_train = sc_x.fit_transform(x_train)
x_test = sc_x.transform(x_test)
```

```python
print (x_train)
```

```python
from sklearn.naive_bayes import GaussianNB #memanggil library naive bayes
classifier = GaussianNB()
classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_test) #menghitung nilai prediksi
```

```python
from sklearn.metrics import confusion_matrix #membuat matriks
cm = confusion_matrix(y_test, y_pred)
print (cm)
```

```python
from matplotlib.colors import ListedColormap #membuat grafik
x_set, y_set = x_train, y_train
x1, x2 = np.meshgrid(np.arange(start = x_set[:, 0].min() - 1, stop = x_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = x_set[:, 1].min() - 1, stop = x_set[:, 1].max() + 1, step = 0.01))
plt.contourf(x1, x2, classifier.predict(np.array([x1.ravel(), x2.ravel()]).T).reshape(x1.shape),
            alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(x1.min(), x1.max())
plt.ylim(x2.min(), x2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(x_set[y_set == j, 0], x_set[y_set == j, 1],
                c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Klasifikasi Data dengan Naive Bayes (Data Training)')
plt.xlabel('Umur')
plt.ylabel('Pendapatan Tahunan Pelanggan')
plt.legend()
plt.show()
```

```python
from matplotlib.colors import ListedColormap #membuat grafik
x_set, y_set = x_test, y_test
x1, x2 = np.meshgrid(np.arange(start = x_set[:, 0].min() - 1, stop = x_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = x_set[:, 1].min() - 1, stop = x_set[:, 1].max() + 1, step = 0.01))
plt.contourf(x1, x2, classifier.predict(np.array([x1.ravel(), x2.ravel()]).T).reshape(x1.shape),
            alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(x1.min(), x1.max())
plt.ylim(x2.min(), x2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(x_set[y_set == j, 0], x_set[y_set == j, 1],
                c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Klasifikasi Data dengan Naive Bayes (Data Testing)')
plt.xlabel('Umur')
plt.ylabel('Pendapatan Tahunan Pelanggan')
plt.legend()
plt.show()
```
